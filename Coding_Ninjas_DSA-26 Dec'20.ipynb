**1.1) Factorial function using Recursion**

```python
import sys
sys.setrecursionlimit(4000)

n = int(input())

def factorial(n):
    if n == 0:                       # Base case for n = 0, to avoid reaching recursion depth limit
        return 1
#     if n > 0:                      # Commented out this step since it is not really required
    return n*factorial(n-1)          # Inductive step for the hypothesis factorial(n-1)

print(factorial(n))
```

**1.2) Sum of first n Natural numbers**

```python
n = int(input())

def sum_(n):
    if n == 0:                       # Base case for n = 0, to avoid reaching recursion depth limit
        return 0

    return n+sum_(n-1)                # Inductive step for the hypothesis sum(n-1)

print(sum_(n))
```

**1.3) Power of a Number**

```python
x, n = input().split()
x = int(x)
n = int(n)

def power(x, n):
    if x == 0:
        return 0

    if n == 0:                            # Base case for n = 0, to avoid reaching recursion depth limit
        return 1

    return x*power(x, n-1)                # Inductive step for the hypothesis sum(n-1)

print(power(x, n))
```

**2.5) Recursive Binary Search**

```python
ar = [int(x) for x in input().split()]
ar.sort()              # Since Binary Search requires only a Sorted array
m = int(input())
n = len(ar)

def BinarySearch(arr, x, si, ei):
    if si > ei:
        return -1

    mid = (si+ei)//2

    if arr[mid]==x:
        return mid
    elif arr[mid]<x:
        return BinarySearch(arr, x, mid+1, ei)
    else:
        return BinarySearch(arr, x, si, mid-1)

print(BinarySearch(ar, m, 0, n-1))
```

**2.6) Recursive Merge Sort**

```python
a = [int(x) for x in input().split()]

def merge(arr1, arr2, a):
    n1 = len(arr1)
    n2 = len(arr2)
    i = 0
    j = 0
    k = 0
    while i < n1 and j < n2:
        if arr1[i] > arr2[j]:
            a[k] = arr2[j]
            j += 1
            k += 1
        else:
            a[k] = arr1[i]
            i += 1
            k += 1

    while j < n2:
        a[k] = arr2[j]
        j += 1
        k += 1

    while i < n1:
        a[k] = arr1[i]
        i += 1
        k += 1

def mergesort(a):
    if len(a) == 0 or len(a) == 1:
        return

    mid = len(a)//2
    a1 = a[0:mid]
    a2 = a[mid:]

    mergesort(a1)
    mergesort(a2)

    merge(a1, a2, a)

mergesort(a)
print(a)
```

**2.7) Quick Sort**

```python
a = [int(x) for x in input().split()]

def partition(a, si, ei):
    pivot = a[si]
    if si > ei:
        return -1
    c = 0
    for i in range(si, ei+1):
        if a[i] < pivot:
            c += 1
    a[si+c], a[si] = a[si], a[si+c]

    piv_ind = si+c

    i = si
    j = ei

    while i < j:
        if (a[i] < pivot):
            i += 1
        elif a[j] >= pivot:
            j -= 1
        else:
            a[i], a[j] = a[j], a[i]
            i += 1
            j -= 1
    return piv_ind


def quickSort(a, si, ei):
    if si >= ei:
        return
    piv_ind = partition(a, si, ei)
    quickSort(a, si, piv_ind-1)
    quickSort(a, piv_ind+1, ei)

quickSort(a, 0, len(a)-1)
print(a)
```

**2.8) Tower of Hanoi**

```python
def towerOfHanoi(n, a, b, c):
    if n == 1:
        print('Move disk 1 from', a, 'to', c)
        return
    towerOfHanoi(n-1, a, c, b)
    print('Move disk', n, 'from', a, 'to', c)
    towerOfHanoi(n-1, b, a, c)

towerOfHanoi(4, 's', 'h', 'd')
```

```python
# # Linked List Opeartion Codes:

class Node:                       # This class contains all the Nodes or Elements from our Linked List
    def __init__(self, data):
        self.data = data
        self.next = None          # By default we set the Reference to next Node as None

"""--------------------------------------------------------------------------------------------------------------------------"""

# Taking the input of the List and converting it to a Singly Linked List in O(n)-
def takeInput():                  # This will take the list input and return the Head
    inputList = [int(ele) for ele in input().split()]

    head = None
    tail = None
    for currData in inputList:
        if currData == -1:
            break

        newNode = Node(currData)

        if head is None:
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = newNode

    return head

"""--------------------------------------------------------------------------------------------------------------------------"""

# Printing the Linked List in O(n)-
def printLL(head):                 # This will print the Linked list by traversing from the Head of the Linked list
    while head is not None:
        print(str(head.data) + ' -> ', end = '')
        head = head.next
    print('None')

"""--------------------------------------------------------------------------------------------------------------------------"""

# Finding the length of the Linked List in O(n)-
def length(head) :                 # This will returns the length of the Linked list
    length = 0

    while head != None:
        length += 1
        head = head.next

    return length

"""--------------------------------------------------------------------------------------------------------------------------"""

# Reversing the links in a Singly Linked List recursively in O(n^2)-
def reverseLLOn2(head):
    if head == None or head.next == None:
        return head

    revhead = reverseLLOn2(head.next)

    curr = revhead
    while curr.next is not None:
        curr = curr.next
    curr.next = head
    head.next = None

    return revhead

"""--------------------------------------------------------------------------------------------------------------------------"""

# Reversing the links in a Singly Linked List recursively in O(n)-
def reverseinOn(head):
    if head is None or head.next is None:
        return head

    smallHead = reverseinOn(head.next)
    tail = head.next
    tail.next = head
    head.next = None

    return smallHead

"""--------------------------------------------------------------------------------------------------------------------------"""

# Reversing the links in a Singly Linked List iteratively in O(n)-
def reverseiter(head):
    if head is None:
        return

    prev = None
    curr = head

    while curr is not None:
        nxt = curr.next
        curr.next = prev
        prev = curr
        curr = nxt

    head = prev
    return head

"""--------------------------------------------------------------------------------------------------------------------------"""

# Finding the Midpoint of a Singly Linked List iteratively in O(n) but requires finding the length first and then iterating
# over the list which overall takes 2 passes-
def midptLL(head):
    if head is None:
        return

    nl = length(head)

    curr = head
    for x in range((nl-1)//2):
        curr = curr.next

    return curr.data

"""--------------------------------------------------------------------------------------------------------------------------"""

# Finding the Midpoint of a Singly Linked List iteratively in O(n)in a single pass-
def midptLL_1_pass(head):
    if head is None:
        return

    slow = head    # This pointer moves an unit step at a time
    fast = head    # This pointer moves at twice speed than slow i.e. takes 2 steps
    while fast.next is not None and fast.next.next is not None:
        slow = slow.next
        fast = fast.next.next

    return slow.data

"""--------------------------------------------------------------------------------------------------------------------------"""

# Check for a Palindromic Linked List (A brilliant solution)-
def isPalindrome(head) :
    #Your code goes here
    if head == None:
        return True

    x, y = '', ''

    while head != None:
        x = x + str(head.data)
        y = str(head.data) + y
        head = head.next

    return (x==y)

"""--------------------------------------------------------------------------------------------------------------------------"""
"""--------------------------------------------------------------------------------------------------------------------------"""


original_head = takeInput()
printLL(original_head)

reverse_list = reverseLL(original_head)
printLL(reverse_list)

reverse_listOn = reverseinOn(original_head)
printLL(reverse_listOn)

reverse_list_iter = reverseiter(original_head)
printLL(reverse_list_iter)

midpt = midptLL(original_head)
print(midpt)

midpt1pass = midptLL_1_pass(original_head)
print(midpt1pass)
```

```python
class Stack:

    def __init__(self):
        self.__data = []         # This is our main array in which the values would be stored

    def push(self, item):
        if type(item) == list:
            self.__data.extend(item)
        else:
            self.__data.append(item)

    def pop(self):
        if self.isempty():
            return
        return self.__data.pop()

    def top(self):
        if self.isempty():
            return
        return self.__data[-1]

    def size(self):
        return len(self.__data)

    def isempty(self):
        return self.size() == 0


s = Stack()
s.push(12)
s.push(13)
s.push([14,15,16])

while not s.isempty():
    print(s.pop(), end=' ')
    print(s.size(), end = ' ')
    print(s.top())
```

```python
class Node :

    def __init__(self, data) :
        self.data = data
        self.next = None


class Stack :

    def __init__(self):
        self.__count = 0         # This is supposed to fetch the length/size of the stack
        self.__head = None

    '''----------------- Public Functions of Stack -----------------'''

    def getSize(self) :
        return self.__count


    def isEmpty(self) :
        return (self.getSize() == 0)


    def push(self, data) :
        pushelem = Node(data)
        pushelem.next = self.__head
        self.__head = pushelem
        self.__count = self.__count + 1


    def pop(self) :
        if self.isEmpty():
            return
        popped = self.__head.data
        self.__head = self.__head.next
        self.__count = self.__count - 1
        return popped


    def top(self) :
        if self.isEmpty():
            return
        return self.__head.data

s2 = Stack()
s2.push(12)
s2.push(13)
s2.push(14)
s2.push(15)
s2.push(16)

print(s2.getSize())
print(s2.isEmpty())
print(s2.top())
while not s2.isEmpty():
    print(s2.pop())
```

### **11) Queues:-**

**11.1) Queue using the Array:**

```python
class Queue:

    def __init__(self):
        self.__data = []                  # This is our main array in which the values would be stored
        self.__count = 0                  # This is supposed to fetch the length/size of the queue
        self.__front = 0                  # This maintains the counter to the front of the queue

    def enqueue(self, item):              # Insertion at the end of the array in O(1)
        if type(item) == list:
            self.__data.extend(item)
            self.__count += len(item)
        else:
            self.__data.append(item)
            self.__count += 1

    def dequeue(self):                    # Removing from the front of the array in O(1)
        if self.isempty():
            return
        popped = self.__data[self.__front]
        self.__front += 1
        self.__count -= 1
        return popped

    def front(self):                      # Extracting the front element of the array in O(1)
        if self.isempty():
            return
        return self.__data[self.__front]

    def size(self):
        return self.__count

    def isempty(self):
        return self.__count == 0


s = Queue()
s.enqueue(12)
s.enqueue(13)
s.enqueue([14,15,16])

while not s.isempty():
    print(s.dequeue(), end=' ')
    print(s.size(), end = ' ')
    print(s.front())
```

**11.2) Queue using the Linked List:**

```python
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None

class Queue :
    def __init__(self):
        self.__count = 0         # This is supposed to fetch the length/size of the queue
        self.__head = None
        self.__tail = None

    '''----------------- Public Functions of Queue -----------------'''

    def getSize(self) :
        return self.__count


    def isEmpty(self) :
        return (self.getSize() == 0)


    def enqueue(self, data):
        enqueueelem = Node(data)
        if self.__head is None:
            self.__head = enqueueelem
            self.__tail = enqueueelem
        else:
            self.__tail.next = enqueueelem
            self.__tail = enqueueelem
        self.__count = self.__count + 1


    def dequeue(self) :
        if self.isEmpty():
            return
        popped = self.__head.data
        self.__head = self.__head.next
        self.__count = self.__count - 1
        return popped


    def front(self) :
        if self.isEmpty():
            return
        return self.__head.data

sq = Queue()
sq.enqueue(12)
sq.enqueue(13)
sq.enqueue(14)
sq.enqueue(15)
sq.enqueue(16)

print(sq.getSize())
print('Queue Empty?: ', sq.isEmpty())
print('The First inserted element in this queue is: ', sq.front())
print()
print('Dequeuing the Queue now:')
while not sq.isEmpty():
    print(sq.dequeue())
print('Queue Empty?: ', sq.isEmpty())
```

**11.3) Queue using two Stacks:**

```python
class QueueUsingTwoStacks:

    def __init__(self):
        self.__s1 = []
        self.__s2 = []

    def enqueue(self, item):

        while len(self.__s1) != 0:
            self.__s2.append(self.__s1.pop())

        self.__s1.append(item)

        while len(self.__s2) != 0:
            self.__s1.append(self.__s2.pop())

    def dequeue(self):
        if self.isempty():
            return
        return self.__s1.pop()

    def front(self):
        if self.isempty():
            return
        return self.__s1[-1]

    def size(self):
        return len(self.__s1)

    def isempty(self):
        return (self.size() == 0)


sq = QueueUsingTwoStacks()
sq.enqueue(12)
sq.enqueue(13)
sq.enqueue(14)
sq.enqueue(15)
sq.enqueue(16)

print(sq.size())
print('Queue Empty?: ', sq.isempty())
print('The First inserted element in this queue is: ', sq.front())
print()
print('Dequeuing the Queue now:')
while not sq.isempty():
    print(sq.dequeue())
print('Queue Empty?: ', sq.isempty())
```

### **12) Binary Trees 1 & 2:-**

```python
class BinaryTreeNode:           # Base Class for storing a Node

    def __init__(self, data):   # We store 3 attributes- i) Data, ii) Left Child Address, iii) Right Child Address
        self.data = data
        self.left = None
        self.right = None

"""---------------------------------------------------------------------------------------------------------------------"""

def printTree(root):           # We will print the Tree showing each of the nodes along with their left and right children
    if root is None:                          # We return with leaf node as a base case
        return

    print(root.data, ':', end=' ')              # We print the root data first
    if root.left:                             # We check if the left child exists
        print('L',root.left.data, end=', ')     # Now we print the left child data
    if root.right:                # We check if the right child exists
        print('R',root.right.data, end='')      # Now we print the right child data
    print()                                   # We peint a new line after printing children of each node

    printTree(root.left)  # Recursive call to the left subtree
    printTree(root.right) # Recursive call to the right subtree

"""---------------------------------------------------------------------------------------------------------------------"""

def treeInput():    # Recursive program to take the Input
    rootdata = int(input())         # We first ask for the root's data

    if rootdata == -1:              # Enter -1 to specify no node
        return None

    root = BinaryTreeNode(rootdata) # We create the node from root ddata

    leftchild = treeInput()         # Recursive call to take the left child input
    rightchild = treeInput()        # Recursive call to take the right child input

    root.left = leftchild           # We assign the left child to the left of the root
    root.right = rightchild         # We assign the right child to the right of the root

    return root

    '''Understand how the input would be given: We first give root, then give input for it's entire left subtree before
       we input the right subtree. For the leaf nodes, it is important to input -1 wherever there is no child node. This
       acts as an important break condition.
       e.g. For a levelwise tree 1 2 3 4 5 -1 7;
       We will input 1 first, then go to 2 then go to its left child 4. Now, 4 is a leaf node, thus its children are -1 -1.
       Now we come to right child of 2 which is 5. Again, 5 is a leaf node too. Thus, we input its children -1 -1. Now we go
       to the right child of our main root which is 3. It now has no left child, thus input -1. Input right child as 7.
       Finally, since 7 is also a leaf node, input -1 -1.
       Thus, the entire input is: 1 2 4 -1 -1 5 -1 -1 3 -1 7 -1 -1 '''

"""---------------------------------------------------------------------------------------------------------------------"""

def treenumNodes(root):  # Recursive program to find the number of nodes in the tree
    if root is None:     # We return 0 as we reach the leaf node
        return 0

    # Now we make recursive calls to the left as well ass the right subtree
    return 1+treenumNodes(root.left)+treenumNodes(root.right)   # 1 is added to ensure we get the addition

"""---------------------------------------------------------------------------------------------------------------------"""

def getSum(root):        # Recursive program to find the sum of nodes in the tree
    if root is None:     # We return 0 as we reach the leaf node
        return 0

    # We call the left subtree as well as the right subtree to get its sum
    return root.data+getSum(root.left)+getSum(root.right)   # Observe we add the root.data to make sure we add the data values

"""---------------------------------------------------------------------------------------------------------------------"""

# # We will now cover all the 3 types of Tree traversals-
'''Notice that the base case is same for all which is we return in case of a leaf node where root is None.
   We always print the data of the root node and recursively call the left as well as the right subtree.
   The only changes we do is the order in which root would be printed & left and right would be called.'''

# Pre-order Traversal:
def preOrder(root):
    if root is None:
        return

    print(root.data, end=' ')           # Pre-order: ROOT, LEFT, RIGHT
    preOrder(root.left)
    preOrder(root.right)

# Pre-order Traversal:
def postOrder(root):
    if root is None:
        return

    postOrder(root.left)                # Post-order: LEFT, RIGHT, ROOT
    postOrder(root.right)
    print(root.data, end=' ')

# In-order Traversal:
def inOrder(root):
    if root is None:
        return

    inOrder(root.left)                # In-order: LEFT, ROOT, RIGHT
    print(root.data, end=' ')
    inOrder(root.right)

"""---------------------------------------------------------------------------------------------------------------------"""

def maxNode(root):    # Recursively finding the node with the largest data
    if root is None:       # Base case for the leaf node, we can return any large negative number
        return -1

    return max(root.data, maxNode(root.left), maxNode(root.right))

"""---------------------------------------------------------------------------------------------------------------------"""

def countNodesGreaterThanX(root, x):    # Recursively finding the node with the largest data
    count = 0                 # We keep a Counter intially set to 0
    if root is None:          # Base case for the leaf node
        return 0

    if root.data > x:
        count += 1

    leftsum =  countNodesGreaterThanX(root.left, x)
    rightsum = countNodesGreaterThanX(root.right, x)

    return count + leftsum + rightsum    # Observe we add the Counter count here

"""---------------------------------------------------------------------------------------------------------------------"""

def height(root):    # Recursive function to find the height of a tree
    if root is None:     # We return 0 upon reaching a leaf node
        return 0

    lh = height(root.left)   # Recursively finding the height of the left subtree
    rh = height(root.right)  # Recursively finding the height of the right subtree

    return 1 + max(lh, rh)   # We do 1+ to add the root node

    '''It is interesting to note here that the height of any tree is just root(1) plus the maximum of the heights
       of the left subtree or the right subtree.'''

"""---------------------------------------------------------------------------------------------------------------------"""


def numLeafNodes(root):   # We recursively find the number of leaf nodes in the tree
    if root is None:
        return 0

    if root.left is None and root.right is None:  # Increment the Counter upon reaching a leaf node
        return 1

    lleaf = numLeafNodes(root.left)    # Recursively finding the number of leaf nodes in the left subtree
    rleaf  = numLeafNodes(root.right)  # Recursively finding the number of leaf nodes in the right subtree

    return lleaf + rleaf

"""---------------------------------------------------------------------------------------------------------------------"""

def printatDepthk(root, k, d=1): # We recursively find the and print the nodes at a level k
    if root is None:
        return

    if d == k:
        print(root.data, end=' ')       # Print the root data if it is at the given level k
        return

    printatDepthk(root.left, k, d=d+1)  # Recursively calling the left subtree after incrementing the counter d
    printatDepthk(root.right, k, d=d+1) # Recursively calling the right subtree after incrementing the counter d

"""---------------------------------------------------------------------------------------------------------------------"""

def changeToDepthTree(root, d=0): # We recursively change the values of nodes at each level
    if root is None:
        return

    root.data = d                        # Change the root's data to level

    changeToDepthTree(root.left, d=d+1)  # Recursively calling the left subtree after incrementing the counter d
    changeToDepthTree(root.right, d=d+1) # Recursively calling the right subtree after incrementing the counter d

    return root                          # Return root since we are modifying the root

"""---------------------------------------------------------------------------------------------------------------------"""

def isNodePresent(root, x): # We recursively check for the presence of node x
    if root:                  # If the root is not None
        if root.data == x:
            return True

        y = isNodePresent(root.right, x)        # Recursively checking in the left subtree

        if not y:
            return isNodePresent(root.left, x)  # Recursively checking in the right subtree
        else:
            return y

    '''This is a very interesting program. Observe that we first check in the right subtree. Only if the data is
       not found in the right subtree, we call the left subtree and return the result as shown in the code. '''

"""---------------------------------------------------------------------------------------------------------------------"""

def printNodesWithoutSibling(root) :
    if root is None:
        return

    # We now check if either  of the children is None
    if root.left is None and root.right is not None:   # We check if right child is not None and then print its data
        print(root.right.data, end=' ')
    elif root.left is not None and root.right is None: # We check if left child is not None and then print its data
        print(root.left.data, end=' ')

    printNodesWithoutSibling(root.left)   # Recursively checking in the left subtree
    printNodesWithoutSibling(root.right)  # Recursively checking in the right subtree

"""---------------------------------------------------------------------------------------------------------------------"""

def removeLeafNodes(root): # Recusrive program to remove leaf nodes
    if root is None:         # Base Case
        return None

    if root.left is None and root.right is None:   # If a node is a leaf node, we return ny removing this node
        return None

    # The nest step is the most crucial one. We need to assign None to the children of leaf node's parent
    root.left = removeLeafNodes(root.left)
    root.right = removeLeafNodes(root.right)

    return root

"""---------------------------------------------------------------------------------------------------------------------"""

def mirrorBinaryTree(root):  # Recusrive program to mirror a tree
    if root is None:
        return True

    rr = mirrorBinaryTree(root.right)  # Recursively getting the left subtree
    rl = mirrorBinaryTree(root.left)   # Recursively getting the right subtree

    root.right = rl                    # Attach left subtree to the right of the root
    root.left = rr                     # Attach rightt subtree to the left of the root

    return root                        # It is important to return the root

"""---------------------------------------------------------------------------------------------------------------------"""
    '''A Tree is called as a Balanced tree if the difference between the left subtree height and the right subtree
       height is either 0 or 1 at max. Also, we use Recursive concept here by checking whether the left and the right
       subtree at the root is balanced or not. By returning the height simultaneously, we save computations and thus
       reduce the complexity to O(n)'''

def heightandcheckBalanced(root):
    if root is None:
        return 0, True

    lh, leftbalance = checkBalanced(root.left)
    rh, rightbalance = checkBalanced(root.right)

    h = 1+max(lh+rh)          # Recursively finding the height

    if abs(lh - rh) > 1:      # Checking for the tree balancing condition
        return h, False

    if leftbalance and rightbalance:    # We return True if both the left as well right subtrees are balanced
        return h, True
    else:
        return h, False

"""---------------------------------------------------------------------------------------------------------------------"""
    """Here we write a Recursive program to find the diameter of the tree with an helper function which would return
       both the height as well as the diameter of the tree. The diameter of a tree can be defined as the maximum distance
       between two leaf nodes. Here, the distance is measured in terms of the total number of nodes present along the
       path of the two leaf nodes, including both the leaves.
       Notice that the diameter can be either of the two cases:
           i)   Passing through the main root, in which case it would be (1+left_height+right_height)
           ii)  Lying in the left or the right subtree """

def diahelper(root):    # Recursive helper function which would return height as well as the diameter
    if root is None:
        return 0, 0

    lh, ld = diahelper(root.left)
    rh, rd = diahelper(root.right)

    return 1+max(lh, rh), max(1+lh+rh, ld, rd)

def diameterOfBinaryTree(root) :
    ht, d = diahelper(root)
    return d

"""---------------------------------------------------------------------------------------------------------------------"""

'''Levelorder Traversal: We traverse through each level of the tree in a left-to-right fashion and consider -1 wherever
   no node or child is present. The Levelwise input would be taken iteratively as recursive call won't really be possible'''

import queue
def LevelInput(root):
    q = queue.Queue()

    rootdata = int(input('Provide the Root: '))
    if rootdata == -1:
        return None

    root = BinaryTreeNode(rootdata)

    while not q.empty():
        current_node = q.get()

        leftchilddata = int(input('Provide left child of ', current_node.data))
        if leftchilddata != -1:
            leftchild = BinaryTreeNode(leftchilddata)
            current_node.left = leftchild
            q.put(leftchild)

        rightchilddata = int(input('Provide right child of ', current_node.data))
        if rightchilddata != -1:
            rightchild = BinaryTreeNode(rightchilddata)
            current_node.right = rightchild
            q.put(rightchild)

    return root

"""---------------------------------------------------------------------------------------------------------------------"""

def printLevelWise(root):
    # Your code goes here
    q = queue.Queue()

    if root is None:
        return None
    else:
        q.put(root)

    while not q.empty():
        curr_node = q.get()
        print(curr_node.data,end='')

        if curr_node.left is not None:
            q.put(curr_node.left)
            print(':L:',end='')
            print(curr_node.left.data,end='')
        else:
            print(':L:-1',end='')
        if curr_node.right is not None:
            q.put(curr_node.right)
            print(',R:',end='')
            print(curr_node.right.data)
        else:
            print(',R:-1')

"""---------------------------------------------------------------------------------------------------------------------"""

'''We have already learnt the 3 types of Traversals of a tree: i) Pre-order, ii) In-order, iii) Post-order. We cannot
   build an unique tree given each of these traversals individually. Thus, for building the tree from these traversals,
   we need to consider combinations of - a) Pre-order & In-order or b) Post-order & In-order.

   Case a) Pre-order & In-order: The procedure is as follows-
            i)   First find the root of the tree from preOrder[0].
            ii)  Find this root somewhere in the middle of the In-order.
            iii) Now, all the nodes to the left of the root in In-order are the left subtree nodes. While those on the
                 right belong to the right subtree. Count the length of the left as well as the right subtree (lc and rc).
            iv)  Now, come to the preOrder and start counting after root in the first index, we will get the left subtree
                 and the right subtree.

   Case b) Post-order & In-order: The procedure is as follows-
            i)   First find the root of the tree from postOrder[-1] (Last element).
            ii)  Find this root somewhere in the middle of the In-order.
            iii) Now, all the nodes to the left of the root in In-order are the left subtree nodes. While those on the
                 right belong to the right subtree. Count the length of the left as well as the right subtree (lc and rc).
            iv)  Now, come to the postOrder and start counting before root in the lasst index, we will get the left subtree
                 and the right subtree.

   Now we will buld the tree using Recursion for both the cases.'''


def buildTree(preOrder, inOrder, n) :
    if n == 0 :
        return None

    if len(preOrder) == 0 or len(inOrder) == 0:
        return

    root = preOrder[0]
    lef_len = inOrder.index(root)

    left_subtree = buildTree(preOrder[1:lef_len+1], inOrder[:lef_len], n)
    right_subtree = buildTree(preOrder[lef_len+1:], inOrder[lef_len+1:], n)

    root = BinaryTreeNode(root)

    root.left = left_subtree
    root.right = right_subtree

    return root


def buildTree(postOrder, inOrder, n) :
    if n == 0 :
        return None

    if len(postOrder) == 0 or len(inOrder) == 0:
        return

    ln = len(postOrder)
    root = postOrder[-1]
    lef_len = inOrder.index(root)

    left_subtree = buildTree(postOrder[:lef_len], inOrder[:lef_len], n)
    right_subtree = buildTree(postOrder[lef_len:ln-1], inOrder[lef_len+1:], n)

    root = BinaryTreeNode(root)

    root.left = left_subtree
    root.right = right_subtree

    return root

"""---------------------------------------------------------------------------------------------------------------------"""

def insertDuplicateNode(root):
    if root is None:
        return

    rl = insertDuplicateNode(root.left)
    insertDuplicateNode(root.right)

    rootnew = BinaryTreeNode(root.data)

    root.left = rootnew
    rootnew.left = rl

    return root



"""---------------------------------------------------------------------------------------------------------------------"""
"""---------------------------------------------------------------------------------------------------------------------"""




# Main

print('Input the Tree first: \n')
root = treeInput()
print()

print('The Tree input given is: \n')
printTree(root)
print()

print('The number of nodes in the tree are: \n')
print(treenumNodes(root))
print()

print('The sum of the nodes in the tree are: \n')
print(getSum(root))
print()

print('The Pre-order traversal of the tree is: \n')
print(preOrder(root))
print()

print('The Post-order traversal of the tree is: \n')
print(postOrder(root))
print()

print('The In-order traversal of the tree is: \n')
print(inOrder(root))
print()

print('The largest node in the tree is: \n')
print(maxNode(root))
print()

print('The height of the tree is: \n')
print(height(root))
print()

print('The number of leaf nodes in the tree are: \n')
print(numLeafNodes(root))
print()

print('The number of nodes greater than 2 in the tree are: \n')
print(countNodesGreaterThanX(root, 0))
print()

print('The nodes at level k are: \n')
print(printatDepthk(root, 1))
print(printatDepthk(root, 2))
print(printatDepthk(root, 3))
print(printatDepthk(root, 4))
print()

print('We check if the node 2 is present in the tree?: \n')
print(isNodePresent(root, 2))
print('We check if the node 9 is present in the tree?: \n')
print(isNodePresent(root, 9))
print()

print('The number of nodes without any siblings are: \n')
print(printNodesWithoutSibling(root))
print()

print('The Diameter of the Tree is: \n')
print(diameterOfBinaryTree(root))
print()




# We are always working with the below tree:
#                        1
#                      /   \
#                     2     3
#                    / \     \
#                   4   5     7
#
#    Notice that 3 has no left child.
```
